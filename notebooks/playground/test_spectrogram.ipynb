```python
import torchaudio
import matplotlib.pyplot as plt

signal, fs = torchaudio.load('output.wav')
signal = signal.mean(axis=0, keepdims=True).numpy().ravel()
signal = signal[58*fs:62*fs]
fig, ax = plt.subplots(figsize=(20, 5))
ax.specgram(signal, Fs=fs, NFFT=1024, noverlap=512, cmap='viridis');
```
